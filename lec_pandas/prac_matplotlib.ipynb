```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig=plt.figure(figsize=(5,5))
ax=plt.axes()
```

```python
x=[105,210,301,440,500]
y=[10,20,30,50,60]

fig=plt.figure(figsize=(5,5))
ax=plt.axes()

ax.plot(x,y)
plt.show()
```

```python
x=[105,210,301,440,500]
y=[10,20,30,50,60]

fig=plt.figure(figsize=(5,5))
ax=plt.axes()

ax.set_title('practice matplotlib')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid()

ax.plot(x,y, marker="o", color="red" ,linestyle='--', linewidth=2.0)
plt.show()
```

```python
x=['A','B','C','D']
y=[10,20,30,50]


fig=plt.figure(figsize=(5,5))
ax=plt.axes()

ax.bar(x,y)
plt.show()
```

```python
x=np.random.randn(1000)

fig=plt.figure(figsize=(5,5))
ax=plt.axes()

ax.hist(x, bins=50)
plt.show()
```

```python
x=np.random.randint(1,101,200)
y=np.random.randint(1,101,200)

fig=plt.figure(figsize=(5,5))
ax=plt.axes()

ax.scatter(x,y)
plt.show()
```

```python
fig=plt.figure()

ax1=fig.add_subplot(221)
ax2=fig.add_subplot(222)
ax3=fig.add_subplot(223)
ax4=fig.add_subplot(224)

plt.subplots_adjust(wspace=0.5, hspace=0.5)
```

```python
fig=plt.figure(figsize=(10,10))

ax1=fig.add_subplot(221)
ax2=fig.add_subplot(222)
ax3=fig.add_subplot(223)
ax4=fig.add_subplot(224)

plt.subplots_adjust(wspace=0.5, hspace=0.5)


#ax1
ax1.set_title('ax1')
x=[105,210,301,440,500]
y=[10,20,30,50,60]
ax1.plot(x,y)

#ax2
x=['A','B','C','D']
y=[10,20,30,50]
ax2.bar(x,y)

#ax3
x=np.random.randn(1000)
ax3.hist(x, bins=50)

#ax4
x=np.random.randint(1,101,200)
y=np.random.randint(1,101,200)
ax4.scatter(x,y)

plt.show()
```
